```python
import numpy as np
import pandas as pd
import pandas_market_calendars as mcal

A_10asset_10=np.load('../results/A_10assetnew_10_obs_10L_10M.npy')
A_10asset_20=np.load('../results/A_10assetnew_20_obs_10L_10M.npy')
A_10asset_30=np.load('../results/A_10assetnew_30_obs_10L_10M.npy')

nyse = mcal.get_calendar('NYSE')
all_trade_days_2019april = nyse.valid_days(start_date='2019-04-01', end_date='2020-12-07')
```

```python
import matplotlib.pyplot as plt
from matplotlib import gridspec
%matplotlib inline

plt.rcParams['figure.figsize']=[18,5]

plt.subplots_adjust(wspace=5)

#gs = gridspec.GridSpec(ncols=3, nrows=2, width_ratios=[1, 1]) 
gs = gridspec.GridSpec(ncols=3, nrows=2, height_ratios=[2,1]) 

fig,axs=plt.subplots(1,3)

ax0 = plt.subplot(gs[0])
ax0.plot(np.mean(A_10asset_10, axis=0), linewidth=2, label='target 10%')
ax0.fill_between([*range(0,425)], np.mean(A_10asset_10, axis=0)-np.std(A_10asset_10, axis=0), np.mean(A_10asset_10, axis=0)+np.std(A_10asset_10, axis=0), color='b', alpha=.1)
ax0.plot(np.mean(A_10asset_30, axis=0), linewidth=2, label='target 20%')
ax0.fill_between([*range(0,425)], np.mean(A_10asset_30, axis=0)-np.std(A_10asset_30, axis=0), np.mean(A_10asset_30, axis=0)+np.std(A_10asset_30, axis=0), color='r', alpha=.1)
ax0.plot(np.mean(A_10asset_20, axis=0), linewidth=2, label='target 30%')
ax0.fill_between([*range(0,425)], np.mean(A_10asset_20, axis=0)-np.std(A_10asset_20, axis=0), np.mean(A_10asset_20, axis=0)+np.std(A_10asset_20, axis=0), color='g', alpha=.1)
ax0.set_ylabel('risk preference')
ax0.tick_params(axis=u'both', which=u'both',length=0)
ax0.set_xticks(np.arange(0,425))
#ax0.grid(which = 'major')
x_ticks = [str(e.date().year)+'/'+str(e.date().month) for e in all_trade_days_2019april]
ax0.set_xticklabels(x_ticks)
#temp = ax0.xaxis.get_ticklabels()
for label in ax0.xaxis.get_ticklabels():
    label.set_visible(False)
for label in ax0.xaxis.get_ticklabels()[::100]:
    label.set_visible(True)

ax0.legend(loc="upper right")
ax0.title.set_text('10 assets')
```
